# G-MOOCSについて
IoTスクールではG-MOOCSと呼ばれるオンデマンド学習システムを使います。
G-MOOCSへの自己登録，コース登録について説明します。

## 自己登録について
以下の手順従い，G-MOOCSへの自己登録をしましょう。
- [こちら](https://expert.idsc-gunma.jp/)にアクセスしてください。
- 「新しいアカウントを作成する」をクリックしてください。

Artboard 1.png

- 以下の画面で，ユーザー名，パスワード，メールアドレス，氏名(姓と名)を記載し「私の新しいアカウントを作成する」をクリックする。
 - **ユーザー名**：G-Moocsにログインするときに必要なユーザー名です。ユーザ名には半角英数小文字 (文字および数字)、アンダースコア (_)、ハイフン (-)、ピリオド (.) またはアットマーク (@) のみ含むことができます。例えば「群大　太郎」というユーザーであれば"gundai-taro"などと入力します。
 - **パスワード**：パスワードには少なくとも半角 8 文字使用する必要があります。
パスワードには少なくとも半角 1 文字の小文字を使用する必要があります。
パスワードには少なくとも半角 1 文字の「*」「 -」「#」のような特殊文字を使用する必要があります。

- 以下のような画面になるので「続ける」をクリックし，メールが受信できていることを確認してください。

- 下のように迷惑メールに分類されている可能性もありますので，確認してください。

- メールのリンク先をクリックすると以下のように登録が承認されます。

## コース登録について
自己登録が済んだら，コースの登録をしましょう。
- ログイン後，画面の右上の「コース」から「2023年IoTスクール」(TA用のテキストは「IoTスクールTA用のテキスト」)をクリックしましょう。

Screenshot 2023-04-13 at 10.41.22.png

- 別途お知らせしている「登録キー」を入力し，「私を登録する」をクリックすることでコース登録が完了します。
