# G-MOOCSについて
IoTスクールではG-MOOCSと呼ばれるオンデマンド学習システムを使います。
G-MOOCSへの自己登録，コース登録について説明します。

## 自己登録について
以下の手順従い，G-MOOCSへの自己登録をしましょう。
- [こちら](https://expert.idsc-gunma.jp/)にアクセスしてください。
- 下の図のように右上に表示されている「新しいアカウントを作成する」をクリックしてください。

Artboard 1.png

- 以下の画面で，ユーザー名，パスワード，メールアドレス，氏名(姓と名)を記載し「私の新しいアカウントを作成する」をクリックする。



Screenshot 2023-04-13 at 9.49.47.png

- 以下のような画面になるので「続ける」をクリックし，群馬大学リカレントMoocsからメールが受信できていることを確認してください。

Screenshot 2023-04-13 at 9.50.04.png

- 下のように迷惑メールに分類されている可能性もありますので，受信トレイにない場合は迷惑メールを確認してください。

Screenshot 2023-04-13 at 9.50.42.png

- メールのリンク先をクリックすると以下のように登録が承認され，自己登録は完了です。

Screenshot 2023-04-13 at 9.51.12.png

---
## コース登録について
自己登録が済んだら，コースの登録をしましょう。
- ログイン後，画面の右上の「コース」から「2023年IoTスクール」(TA用のテキストは「IoTスクールTA用のテキスト」)をクリックしましょう。

Screenshot 2023-04-13 at 10.41.22.png

- 別途お知らせしている「登録キー」を入力し，「私を登録する」をクリックすることでコース登録が完了します。

Screenshot 2023-04-13 at 10.48.27.png
