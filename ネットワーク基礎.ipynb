# ネットワーク基礎

---
Wi-FiのSSIDを調べるには
```
iwgetid
```
設定ファイルは
```
/etc/wpa_supplicant/wpa_supplicant.conf
```
設定を変更するには
```
sudo raspi-config
```
→1 SystemOptions → S1 Wireless LAN

---
自分のIPアドレス，MACアドレスを調べるコマンド
```
ip a
```

スクリーンショット 2024-12-25 17.09.03.png

設定ファイルは
```
/etc/dhcpcd.conf
```

---
FQDN(Fully Qualified Domain Name)を調べるコマンド

```
hostname -f
```
設定ファイルは\
FQDNの一部（ホスト名部分）
```
/etc/hostname
```

ホスト名をFQDNに対応付け
```
/etc/hosts
```

設定を変更するには
```
sudo raspi-config
```
→1 SystemOptions → S4 Hostname

ホスト名には以下の制約がある：
-	英数字（a-z, 0-9）とハイフン（-）のみ使用可能。
-	ハイフンで始めたり終わったりしない。
